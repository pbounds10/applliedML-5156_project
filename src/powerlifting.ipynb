### Function to show the weights of the linear models

```python
# Taken from Lab3 Writen by Dr. Minwoo Lee for ITCS 5156
'''
Dr. Minwoo Lee, "Lab 3: Linear Models", ITCS 5156: Applied Machine Learning, 9/10/2022, 
    University of North Carolina at Charlotte.
'''
def autolabel(ax, rects):
    """ Attach a text label above each bar in *rects*, displaying its height.
    
        Args:
            ax (matplotlib.Axes): Axe output by matplotlib 
                See https://matplotlib.org/stable/api/axes_api.html
            
            rects (matplotlib.container.BarContainer): Container for bars in bar plots.
                See https://matplotlib.org/stable/api/container_api.html#matplotlib.container.BarContainer
    """
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:0.3f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', color='blue')

def show_weights(model, names):
    
    # Combine both the coefficients and intercept to present
    w = np.append(model.coef_, model.intercept_)
    # Make sure names is a list and append the missing bias label
    # to names.
    names = list(names) + ['bias/intercept']

    plt.figure(figsize=(12,3))
    
    # TODO 8.1
    rects = plt.bar(names,height=w,color='springgreen')
    
    ax = plt.gca()
    ax.set_xticks(range(len(w)))
    ax.set_xticklabels(names, rotation = 90)
    # TODO 8.2
    autolabel(ax,rects)
```

### Now write a function for testing linear regression models (or I guess any model but go this one linear)

 Truong et al 2019 uses Root Mean Squared Logarithmic Error. So from sklearn lets bring in MSLE and take the root

```python
from sklearn.metrics import mean_squared_log_error

def linmodeltester(model,data):
    '''
    inputs:
    model - the ML model

    data - dict with: X_train, X_test, t_train, t_test
    '''

    model.fit(data['X_train'],data['t_train'])

    train_score =  model.score(data['X_train'],data['t_train'])

    test_score = model.score(data['X_test'],data['t_test'])

    y_test = model.predict(data['X_test'])

    y_train = model.predict(data['X_train'])

    print('The training score is: ', train_score)
    print('The test score is: ', test_score)
    print()

    if (min(y_test)<0):
        print('The predicted test values have negative numbers: ',np.sum(np.array(y_test) < 0, axis=0))
        print('Out of the total ',len(y_test))
        print('The percentage of the data is: ', np.sum(np.array(y_test) < 0, axis=0)/len(y_test))
        print('setting these values to 0')
        y_test[y_test<0] = 0
        print()

    if (min(y_train)<0):
        print('The predicted train values have negative numbers: ',np.sum(np.array(y_train) < 0, axis=0))
        print('Out of the total ',len(y_train))
        print('The percentage of the data is: ', np.sum(np.array(y_train) < 0, axis=0)/len(y_train))
        print('setting these values to 0')
        y_train[y_train<0] = 0
        print()


    #take the square root to compare to the paper
    rmsle_test = np.sqrt(mean_squared_log_error(t_test, y_test))
    rmsle_train = np.sqrt(mean_squared_log_error(t_train, y_train))

    print('The training RMSLE is: ', rmsle_train)
    print('The test RMSLE is: ', rmsle_test)

    show_weights(model,data['X_test'].columns.values)

    plt.figure(figsize=(25,10))
    plt.subplot(1,2,1)
    plt.scatter(y_test,t_test)
    plt.title('Testing')
    plt.subplot(1,2,2)
    plt.scatter(y_train,t_train)
    plt.title('Training')

    return test_score, train_score


```
