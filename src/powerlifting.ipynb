### Copy the data so you don't have to reimport. Then drop any column that gives info that isn't needed like Names, Federations, Etc

```python
df = power.copy()

display(df.describe())

display(df.head(10))

display(df.columns)

#drop columns that don't tell you anything

df.drop(['Name', 'Country', 'State', 'Federation', 'Division','Event', 'Place',
       'AgeClass','WeightClassKg',
       'ParentFederation', 'Date', 'MeetCountry', 'MeetState', 'MeetTown',
       'MeetName','Dots', 'Wilks', 'Glossbrenner','BirthYearClass','Goodlift'],axis=1,inplace=True)


```

### Convert Sex to Female = False and Male = True

```python
df['Tested'].fillna(False, inplace = True)
df['Tested'].replace(to_replace='Yes',value=True,inplace=True)
df['Sex'].replace(to_replace='Mx',value=True,inplace=True)
df['Sex'].replace(to_replace='M',value=True,inplace=True)
df['Sex'].replace(to_replace='F',value=False,inplace=True)

display(df['Tested'].unique())
display(df['Tested'].dtypes)
display(df['Sex'].unique())
display(df['Sex'].dtypes)
```

### Remove any rows that are missing:
1. Age
2. Body Weight
3. Total Score

### Then replace any best score that doesn't have a clean attempt

```python
df.columns[df.isna().any()].tolist()

df = df[df['TotalKg'].notna()]
df = df[df['Age'].notna()]
df = df[df['BodyweightKg'].notna()]


exc_names = ['Bench','Squat','Deadlift']

exc_cols = np.concatenate((df.columns[df.columns.str.contains(exc_names[0])],
                            df.columns[df.columns.str.contains(exc_names[1])],
                            df.columns[df.columns.str.contains(exc_names[2])]))

print(exc_cols)

#fill attempt with 0 if they are nan
df[exc_cols] = df[exc_cols].fillna(0)

#drop all other nan rows
df.dropna(inplace=True)

#replace the best attempt of a lift who failed all lifts with 0
df[df.columns[df.columns.str.contains('Best')]]

display(df.head(10))


```

### Go through each exercise and see if the lifter failed an attempt. If so add this to a new column that documents any failures

```python
df_final = df.copy()

#make a new column to tell if a lifter missed an attempt
df_final['Squat_fail'] = 0
df_final['Deadlift_fail'] = 0
df_final['Bench_fail'] = 0

#check to see if a lifter failed by checking for negative value
for ii,nn in enumerate(exc_names):  

    cols = df_final.columns[df_final.columns.str.contains(nn)]

    df_final[nn+'_fail'] = (df_final[cols]<0).any(axis=1)

display(df_final.describe())

display(df_final.head(10))
```

### Remove the columns contain attempts and only leave the best score for each exercise

```python
#remove the best attempts from the excerise column names
exc_cols = [nn for nn in exc_cols if 'Best' not in nn]

print(exc_cols)

#drop the excerises from the dataframe
df_final.drop(exc_cols,axis=1,inplace=True)

display(df_final.describe())

display(df_final.head(10))
```

### One-Hot-Encode the equipment column

```python
df_equip = pd.get_dummies(df_final['Equipment'])

display(df_equip.head(5))

df_final.drop('Equipment',axis=1,inplace=True)

df_final = pd.concat([df_final,df_equip],axis=1)

display(df_final.head(20))
```

### Convert Boolean to 1 and 0 so that plotting can be done

```python
bool_cols = df_final.loc[:,df_final.dtypes == bool].columns.values

display(bool_cols)

display(df_final[bool_cols])

df_final[bool_cols] = df_final[bool_cols].astype(int)

display(df_final[bool_cols])
```

### Split the data into training and test data

```python
from sklearn.model_selection import train_test_split

df_features = df_final.drop(['TotalKg','Best3SquatKg','Best3BenchKg','Best3DeadliftKg'],axis=1)
df_targets = df_final.loc[:,'TotalKg']

display(df_features.info())
display(df_targets.info())

def data_splitting(data, target):
    
    X_train, X_test, t_train, t_test = train_test_split(data,
                                                        target,
                                                        test_size=0.20,
                                                        random_state=0)

    return X_train, X_test, t_train, t_test

X_train, X_test, t_train, t_test = data_splitting(df_features,df_targets)

display(len(X_train))
```

### Function to show the weights of the linear models

```python
# Taken from Lab3 Writen by Dr. Minwoo Lee for ITCS 5156
def autolabel(ax, rects):
    """ Attach a text label above each bar in *rects*, displaying its height.
    
        Args:
            ax (matplotlib.Axes): Axe output by matplotlib 
                See https://matplotlib.org/stable/api/axes_api.html
            
            rects (matplotlib.container.BarContainer): Container for bars in bar plots.
                See https://matplotlib.org/stable/api/container_api.html#matplotlib.container.BarContainer
    """
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:0.3f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', color='blue')

def show_weights(model, names):
    
    # Combine both the coefficients and intercept to present
    w = np.append(model.coef_, model.intercept_)
    # Make sure names is a list and append the missing bias label
    # to names.
    names = list(names) + ['bias/intercept']

    plt.figure(figsize=(12,3))
    
    # TODO 8.1
    rects = plt.bar(names,height=w,color='springgreen')
    
    ax = plt.gca()
    ax.set_xticks(range(len(w)))
    ax.set_xticklabels(names, rotation = 90)
    # TODO 8.2
    autolabel(ax,rects)
```

### Now train a linear model. will try ridge regression here

```python
num_plot = 5000

from sklearn.linear_model import Ridge

rd = Ridge(alpha=1)

rd.fit(X_train[0:num_plot],t_train[0:num_plot])

train_score =  rd.score(X_train[0:num_plot],t_train[0:num_plot])

test_score = rd.score(X_test[0:num_plot],t_test[0:num_plot])

y_test = rd.predict(X_test[0:num_plot])

print('The training score is: ', train_score)
print('The test score is: ', test_score)


show_weights(rd,df_features.columns.values)


plt.figure()
plt.scatter(range(0,num_plot),y_test[0:num_plot])
plt.scatter(range(0,num_plot),t_test[0:num_plot])
```

### Try Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor


# tr_prune = DecisionTreeRegressor(max_depth=100,random_state=0)
tr_prune = DecisionTreeRegressor(max_depth=None,random_state=0)

def test_branches(model,end_param,steps):

    acc = {'param':[],'test_acc':[],'train_acc':[]}

    for ii in range(2,end_param,steps):

        model.set_params(max_depth=ii)

        model.fit(X_train,t_train)

        train_score =  model.score(X_train,t_train)

        test_score = model.score(X_test,t_test)

        y_test = model.predict(X_test)
        print('Number of Param: ', ii)
        # print('The training score is: ', train_score)
        # print('The test score is: ', test_score)

        acc['param'].append(ii)
        acc['test_acc'].append(test_score)
        acc['train_acc'].append(train_score)

    plt.scatter(acc['param'],acc['test_acc'],label='Test Accuracy')
    plt.scatter(acc['param'],acc['train_acc'],label='Train Accuracy')
    plt.legend(loc='right')

    return acc

acc_200 = test_branches(tr_prune,200,10)
```

### Now try Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(warm_start=True,
                            bootstrap=True,
                            oob_score=True,
                            random_state=0)

def test_trees(model,end_param,steps):

    acc = {'param':[],'test_acc':[],'train_acc':[]}

    for ii in range(2,end_param,steps):

        model.set_params(n_estimators=ii)

        model.fit(X_train,t_train)

        train_score =  model.score(X_train,t_train)

        test_score = model.score(X_test,t_test)

        y_test = model.predict(X_test)
        print('Number of Param: ', ii)
        # print('The training score is: ', train_score)
        # print('The test score is: ', test_score)

        acc['param'].append(ii)
        acc['test_acc'].append(test_score)
        acc['train_acc'].append(train_score)

    plt.scatter(acc['param'],acc['test_acc'],label='Test Accuracy')
    plt.scatter(acc['param'],acc['train_acc'],label='Train Accuracy')
    plt.legend(loc='right')

    return acc

rf_acc_200 = test_trees(rf,50,5)
```
