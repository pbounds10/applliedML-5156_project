```python
import pandas as pd
import numpy as np
```

```python
power = pd.read_csv(r'D:\Desktop\College Fall 2022\Applied AI\project\openpowerlifting-2022-09-09\openpowerlifting.csv')
```

### Copy the data so you don't have to reimport. Then drop any column that gives info that isn't needed like Names, Federations, Etc

```python
df = power.copy()

df['Tested'].fillna("No", inplace = True)

display(df.describe())

display(df.head(10))

display(df.columns)

#drop columns that don't tell you anything

df.drop(['Name', 'Country', 'State', 'Federation', 'Division','Event', 'Place',
       'AgeClass',
       'ParentFederation', 'Date', 'MeetCountry', 'MeetState', 'MeetTown',
       'MeetName','Dots', 'Wilks', 'Glossbrenner','BirthYearClass','Goodlift'],axis=1,inplace=True)


```

```python
display(df.head(10))
```

### Remove any rows that are missing:
1. Age
2. Body Weight
3. Total Score

### Then replace any best score that doesn't have a clean attempt

```python
df.columns[df.isna().any()].tolist()

df = df[df['TotalKg'].notna()]
df = df[df['Age'].notna()]
df = df[df['BodyweightKg'].notna()]


exc_names = ['Bench','Squat','Deadlift']

exc_cols = np.concatenate((df.columns[df.columns.str.contains(exc_names[0])],
                            df.columns[df.columns.str.contains(exc_names[1])],
                            df.columns[df.columns.str.contains(exc_names[2])]))

print(exc_cols)

#fill attempt with 0 if they are nan
df[exc_cols] = df[exc_cols].fillna(0)

#drop all other nan rows
df.dropna(inplace=True)

#replace the best attempt of a lift who failed all lifts with 0
df[df.columns[df.columns.str.contains('Best')]]

display(df.head(10))


```

### Go through each exercise and see if the lifter failed an attempt. If so add this to a new column that documents any failures

```python
df_final = df.copy()

#make a new column to tell if a lifter missed an attempt
df_final['Squat_fail'] = 0
df_final['Deadlift_fail'] = 0
df_final['Bench_fail'] = 0

#check to see if a lifter failed by checking for negative value
for ii,nn in enumerate(exc_names):  

    cols = df_final.columns[df_final.columns.str.contains(nn)]

    df_final[nn+'_fail'] = (df_final[cols]<0).any(axis=1)

display(df_final.describe())

display(df_final.head(10))
```

### Remove the columns contain attempts and only leave the best score for each exercise

```python
#remove the best attempts from the excerise column names
exc_cols = [nn for nn in exc_cols if 'Best' not in nn]

print(exc_cols)

#drop the excerises from the dataframe
df_final.drop(exc_cols,axis=1,inplace=True)

display(df_final.describe())

display(df_final.head(10))
```

### One-Hot-Encode the equipment column

```python
df_equip = pd.get_dummies(df_final['Equipment'])

display(df_equip.head(5))

df_final.drop('Equipment',axis=1,inplace=True)

df_final = pd.concat([df_final,df_equip],axis=1)

display(df_final.head(20))
```
